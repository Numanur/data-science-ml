```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/drive/My Drive/Colab Datasets/heart-disease.csv")
```

```python
df["age"].plot()
```

```python
df["age"].hist()
```

```python
df["chol"].plot()
```

```python
df.plot(x="age", y="chol", kind="scatter", color="red", alpha=1)#alpha means the intensity of colour
```

```python
#Plotting by adding titles
graph = df.plot(x="age", y="chol", kind="bar")
graph.set_title("Age vs Chol")
graph.set_xlabel("Age")
graph.set_ylabel("Cholesterol")
plt.show()
```

```python
subset = df.head(10)
ax = subset.plot(x="age", y="chol", kind="bar")
ax.set_title("Age vs Chol")
ax.set_xlabel("Age")
ax.set_ylabel("Chol")
plt.show()
```
