# **seed function application**


```python
np.random.seed(0)
random1 = np.random.randint(10,20, size=(3,3))
random1
```

```python
#same seed value as previous, so same random num generated
np.random.seed(0)
random2 = np.random.randint(10,20, size=(3,3))
random2
```

```python
#changing the seed value generate different random num
np.random.seed(100)
random3 = np.random.randint(10,20, size=(3,3))
random3
```

So seed value works as a key. same key generates same random number. different key for different number.

# **Some Numpy functions**

```python
print(a2)
print(np.sum(a2))#works faster than python sum() function
print(np.min(a2))
print(np.max(a2))
print(np.mean(a2))
print(np.var(a2))
print(np.std(a2)) # std_dev = sqrt of variance
```

**Variance and Standard Deviation**

Variance means how spread each number from the mean value.
- High variance: wider range of numbers
- Low variance: Lower range of numbers

Standard deviation is the standard spread of the data.

```python
high_var = np.array([1,100,500,1000,5000])
low_var = np.array([2,4,6,8,10])
print(np.var(high_var))
print(np.var(low_var))
print(np.mean(high_var))
print(np.mean(low_var))
print(np.std(high_var))
print(np.std(low_var))
```
